```python
from impactchart.model import XGBoostImpactModel

from impactchart.backend.bokeh_backend import BokehBackend

import pandas as pd
import numpy as np

from bokeh.plotting import show
```

```python
df_data = pd.DataFrame(
    {
        "x1": np.linspace(0, 100, 101),
    }
)

df_data["y"] = (
    np.sin(df_data["x1"] * 8.0 * np.pi / 100.0) + (df_data["x1"] - 50.0) / 25.0
)

df_data
```

```python
X = df_data[["x1"]]
y = df_data["y"]
```

```python
impact_model = XGBoostImpactModel()
```

```python
impact_model.fit(X, y)
```

```python
impact_model.charts(X);
```

```python
bokeh_backend = BokehBackend()
```

```python
charts = impact_model.charts(X, backend=bokeh_backend)
```

```python
show(charts["x1"])
```
